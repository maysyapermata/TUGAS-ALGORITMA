```python
def fibonacci32(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib = [0, 1]
        for i in range(2, n):
            fib.append(fib[i-1] + fib[i-2])
        return fib
    
mintaDigit32 = fibonacci32(9)
print("9 digit fibonacci pertama adalah: ")
print(mintaDigit32)
```
