```python
#Kasus
print("Menentukan bilangan prima pada rentang 1000n s/d 1000(n+1) berdasarkan dua digit terakhir dari NIM.")
print("NIM : 2308541032")
print("Rentang 32000 s/d 33000.")

#Masukkan rentang awal dan akhir
awal_32 = 32000
akhir_32 = 33000

#Definisikan sebuah fungsi untuk menghitung bilangan prima 
def hitung_prima_32(m_32):
    if m_32 <= 1:
        return False
    for p_32 in range (2, int(m_32**0.5)+1):
        if m_32 % p_32 == 0:
            return False
    return True

#Buat perintah agar susunan bilangan prima sesuai dengan rentang awal dan akhir
prima_32 = []
for m_32 in range (awal_32, akhir_32 + 1):
    if hitung_prima_32(m_32):
        prima_32.append(m_32)
        
#Print hasil
print(f'Anggota bilangan prima diantara rentang {awal_32} sampai dengan {akhir_32} adalah')
print(prima_32)
```
