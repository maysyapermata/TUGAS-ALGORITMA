```python
# Persamaan Kuadrat
# import numpy as np
import matplotlib.pyplot as plt
import cmath

#Menentukan akar-akar persamaan kuadrat
print("PROGRAM UNTUK MENENTUKAN AKAR-AKAR DARI PERSAMAAN KUADRAT")
#Langkah 1 : Input nilai a, b dan c untuk persamaan kuadrat
a_32 = float(input("nilai a =")) 
b_32 = float(input("nilai b ="))  
c_32 = float(input("nilai c ="))
print("Persamaan kuadratnya adalah {}x**2+{}x+{}".format(a_32,b_32,c_32))
#Langkah 2 : Hitung diskriminan dengan rumus b**2 - 4ac
D_32 = b_32**2-(4*a_32*c_32)
#Langkah 3 : Hitung akar-akar persamaan kuadrat dengan rumus abc
x1_32 = (-b_32+cmath.sqrt(D_32))/2*a_32  
x2_32 = (-b_32-cmath.sqrt(D_32))/2*a_32  
#Langkah 4 : Print hasil
print(f"Nilai diskriminan dari persamaan kuadrat tersebut adalah D = {D_32}")
print("Akar-akar persamaan kuadrat tersebut adalah ({:.0f},{:.0f})".format(x1_32,x2_32))

print("------------------------------------------------------------------------------------------")

#Menentukan persamaan kuadrat jika yang diketahui akar-akarnya
print("PROGRAM UNTUK MENENTUKAN PERSAMAAN KUADRAT JIKA YANG DIKETAHUI AKAR-AKARNYA")
#Langkah 1 : Input nilai x1 dan x2 sebagai akar dari persamaan kuadrat
x1_32 = float(input("nilai x1 =")) 
x2_32 = float(input("nilai x2 ="))
#Langkah 2 : Hitung nilai a, b dan c berdasarkan informasi yang di dapat
a_32 = 1 / x1_32 - x2_32
b_32 = - (a_32 * (x1_32 + x2_32))
c_32 = a_32 * x1_32 * x2_32
#Langkah 3 : Print hasil
print(f"Persamaan kuadrat: y = {a_32}x^2 + {b_32}x + {c_32}")

print("------------------------------------------------------------------------------------------")

#Menentukan titik puncak dan sumbu simetri
print("PROGRAM UNTUK MENENTUKAN TITIK PUNCAK DAN SUMBU SIMETRI")
#Langkah 1 : Input nilai a, b dan c untuk persamaan kuadrat
a_32 = float(input("nilai a =")) 
b_32 = float(input("nilai b ="))  
c_32 = float(input("nilai c ="))
#Langkah 2 : Masukkan ke dalam rumus
x_puncak = -b_32 / (2 * a_32)
y_puncak = a_32 * x_puncak**2 + b_32 * x_puncak + c_32
sumbu_simetri = f"x = {x_puncak:.2f}"
#Langkah 3 : Print hasil
print(f"Titik puncak: ({x_puncak:.2f}, {y_puncak:.2f})")
print(f"Sumbu simetri: {sumbu_simetri}")

print("------------------------------------------------------------------------------------------")

#Menentukan grafik persamaan kuadrat
print("PROGRAM UNTUK MENENTUKAN grafik persamaan kuadrat")
#Langkah 1 : Input nilai a, b dan c untuk persamaan kuadrat
a_32 = float(input("nilai a =")) 
b_32 = float(input("nilai b ="))  
c_32 = float(input("nilai c ="))
#Langkah 2 : Mendefinisikan fungsi kuadrat
def f(x):
    return a_32 * x**2 + b_32 * x + c_32
#Rentang x
x_min_32 = -5
x_max_32 = 5
num_points = 100
#Hitung x dan y
x_32 = np.linspace(x_min_32, x_max_32, num_points)
y_32 = f(x_32)
#Gambar grafik
plt.plot(x_32, y_32)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Grafik Persamaan Kuadrat")
#Show grafik
plt.show()
```
