```python
### import numpy as np
import matplotlib.pyplot as plt

print("Pertama-tama masukkan titik-titik A,B,C,D yang diketahui")
ax_32 = float(input("Nilai x pertama dari garis pertama ="))
ay_32 = float(input("Nilai y pertama dari garis pertama ="))
bx_32 = float(input("Nilai x kedua dari garis pertama ="))
by_32 = float(input("Nilai y kedua dari garis pertama ="))
print("Masukkan nilai x dan y untuk garis kedua")
cx_32 = float(input("Nilai x pertama dari garis kedua ="))
cy_32 = float(input("Nilai y pertama dari garis kedua ="))
dx_32 = float(input("Nilai x kedua dari garis kedua ="))
dy_32 = float(input("Nilai y kedua dari garis kedua ="))
    
#Selanjutnya dihitung gradien pertama dan kedua dengan rumus y2-y1/x2-x1
m1_32 = (by_32-ay_32)/(bx_32-ax_32)
m2_32 = (dy_32-cy_32)/(dx_32-cx_32)
    
#Lalu akan mencari titik potong pertama dan kedua terhadap sumbu y 
tp1_32 = -m1_32*ax_32+ay_32
tp2_32 = -m2_32*cx_32+cy_32
    
print("Berdasarkan titik-titik A dan B yang diketahui, didapatkan persamaan garis pertama, yaitu y={}x+{}".format(m1_32,tp1_32))
print("Berdasarka titik-titik C dan D yang diketahui, didapatkan persamaan garis pertama, yaitu y={}x+{}".format(m2_32,tp2_32))

#Lalu dibuat semua kemungkinan yang akan terjadi
if m1_32==m2_32:
    print("Kedua garis sejajar, Solusinya akan berjumlah tak terhingga")
    if tp1_32==tp2_32:
        print("Persamaan garis sama, kedua garis berhhimpitan, solusinya tak berhingga.")
        x_32=np.linspace(tp1_32-5,tp1_32+5,100)
        plt.plot(x_32,m1_32*(ax_32)+ay_32,color='blue')
        plt.plot(x_32,m2_32*(cx_32)+cy_32,color='green')
        plt.axvline(x=0,color='black')
        pltaxhline(y=0,color='black')
    else:
        ("Hasil dari kedua garis tidak terdapat penyelesaian")
        x_32=np.linspace(tp1_32-5,tp1_32+5,100)
        plt.plot(x_32,m1_32*(x-ax_32)+ay_32,color='red')
        plt.plot(x_32,m2_32*(x-cx_32)+cy_32,color='purple')
        plt.axvline(x=0,color='black')
        plt.axhline(y=0,color='black')
        plt.show()
else:
    print("Terdapat satu titik potong")
    #Program untuk menentukan titik potong
    tp_x32 = (-m2_32*cx_32+cy_32+m1_32*ax_32-ay_32)/(m1_32-m2_32)
    tp_y32 = m1_32*(tp_x32-ax_32)+ay_32
    print("Titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x32,tp_y32))
    
x=np.linspace(tp_x32-5,tp_x32+5,100)
plt.plot(x,m1_32*(x-ax_32)+ay_32,color='yellow',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m1_32,ax_32,ay_32))
plt.plot(x,m2_32*(x-cx_32)+cy_32,color='pink',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m1_32,cx_32,cy_32))
plt.axvline(x=0,color='black')
plt.axhline(y=0,color='black')
plt.plot(tp_x32,tp_y32,color='green',marker='o',markersize=4)
plt.title('Grafik persamaan garis')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='best')
plt.show()
```
